#MLP model 

**Step 1 :** Import the necessary librairies 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
from keras.optimizers import SGD, RMSprop
import numpy as np
```

**Step 2 :** Load the data 

```python
x_train_url = 'https://raw.githubusercontent.com/cam31415/ML-project-COLAS-/main/x_train.csv'
y_train_url = 'https://raw.githubusercontent.com/cam31415/ML-project-COLAS-/main/y_train.csv'
x_test_url = 'https://raw.githubusercontent.com/cam31415/ML-project-COLAS-/main/x_test.csv'
df_X_train = pd.read_csv(x_train_url)
df_Y_train = pd.read_csv(y_train_url)
df_X_test = pd.read_csv(x_test_url)
```

**Step 3 :** Preprocess the data 

```python
df_X_train.drop('Tmix', inplace=True, axis=1)
df_X_train.drop('m_power', inplace=True, axis=1)
df_X_train.drop('Tosc', inplace=True, axis=1)
df_X_train.drop('Unnamed: 0', inplace=True, axis=1)

df_X_test.drop('Tmix', inplace=True, axis=1)
df_X_test.drop('m_power', inplace=True, axis=1)
df_X_test.drop('Tosc', inplace=True, axis=1)
df_X_test.drop('Unnamed: 0', inplace=True, axis=1)

df_Y_train.drop('Unnamed: 0', inplace=True, axis=1)
df_Y_train['target'] = df_Y_train['target']-1 
```

```python
df_X_test
```

```python
df_Y_train
```

**Step 4 :** Scale the features 

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_X_train)
df_X_test = scaler.fit_transform(df_X_test)
```

**Step 5 :** Split the data into training and test sets 

```python
df_X_train, X_test_other, df_Y_train, y_test = train_test_split(X_scaled, df_Y_train, test_size=0.25, random_state=42)
print("y_train shape", y_test.shape)
```

We chose a test size that is 25% of the total size because we have df_X_test that is 3840 columns and 25% of 15360 is 3840, so that the sizes match 

**Step 6 :** Define the optimizer, build  and compile the mlp model 

```python
optimizer = tf.keras.optimizers.SGD(learning_rate=0.7)

model = Sequential()
model.add(Dense(64, input_dim=df_X_train.shape[1], activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(8, activation='softmax'))


model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

**Step 7 :** Train the model 

```python
history = model.fit(df_X_train, df_Y_train, epochs=100, batch_size=64, verbose=2, validation_data=(X_test_other, y_test))
```

**Step 8 :** Predict 

```python
y_pred_prob = model.predict(df_X_test)
y_pred = np.argmax(y_pred_prob, axis=1)+1 
```

**Step 9 :** Addition of the 'id' column on the test set & concatenate it and the predicted target column

```python
df_test = pd.DataFrame(df_X_test)
df_test['id'] = df_test.index

y_pred_df = pd.DataFrame(y_pred, columns=['target'])
y_pred_df['id'] = df_test['id']
y_pred_df = y_pred_df[['id', 'target']]
```
